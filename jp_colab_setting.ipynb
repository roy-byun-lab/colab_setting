```python
'''
マジックコマンドは、%、%%を使用します。%は1行でコマンドを実行し、%%は複数行のコマンドを実行します。
'''
#ファイルの作成と書き込み
%%writefile  test.py
print('hello world!')
'''셀 명령문 실행(!), cat: 파일 내용 읽기'''
#ファイルの内容確認
!cat test.py

# test.pyを実行する
%run test.py

#ファイルのダウンロード
from google.colab import files
# ブラウザでダウンロードが確認できます。
files.download('test.py')

#クラウドへのファイルのアップロード
myupload = files.upload()

#アップロードされたファイルの確認
!ls
```

```python
# Google ドライブの連携
import os
print(os.getcwd())
!ls

# 実行時に表示されるURLをクリックし、許可すると認証キーが表示されます。そのキーをコピーし、URLの下の空欄に貼り付けると、マウントが成功します。
from google.colab import drive
drive.mount('./MyDrive')

# マウントされた自分のドライブを確認しましょう。
!ls

# そのドライブに移動します。
# 自分のドライブはリモートサーバーではなくローカルサーバーと見なされるため、コマンドを実行する際には先頭に「!」を付けません
cd MyDrive/My Drive

# 自分のドライブの全体リストが表示されます。
ls

# 特定のファイルにアクセスしたい場合は、次のようにアクセスします。
import pandas as pd
df = pd.read_csv("./MyDrive/test/test.csv")
```
